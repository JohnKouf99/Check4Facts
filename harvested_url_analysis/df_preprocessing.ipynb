```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../data/check4facts_harvesting.csv')
df2 = pd.read_csv('../data/cyprus_harvesting.csv')
```

```python
# Step 1: Count unique ids for each claim_id
id_count_per_claim = df1.groupby('claim_id')['id'].nunique()
 
# Step 2: Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(id_count_per_claim, bins=np.arange(id_count_per_claim.min(), id_count_per_claim.max() + 2) - 0.5, edgecolor='black')
plt.title('Histogram of Unique IDs per Claim ID')
plt.xlabel('Number of Unique IDs')
plt.ylabel('Frequency')
plt.xticks(np.arange(id_count_per_claim.min(), id_count_per_claim.max() + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```
