```python
#reading the pickle file
file_path = 'cyprus_list.pkl'

# Deserialize the list from the file using pickle
with open(file_path, 'rb') as file:
    my_list = pickle.load(file)

# Print the list
print(len(my_list))
```

```python
#for each url, we harvest its contents, creating a df 
dfs=[]
for i in range(len(claims_df1)):
    claim = claims_df1[i].strip()
    my_harvester = Harvester(my_list[i], claim, timeout=1000, claim_id=i)
    df = my_harvester.run()
    dfs.append(df)
main_df = pd.concat(dfs, reset_index=True)
```
